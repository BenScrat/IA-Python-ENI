# Les formes d'apprentissage


Les objectifs principaux du Machine Learning sont de produire un modèle entraîné sur des données, ce qui permet d'obtenir un programme capable de faire des prédictions et de prendre des décisions.

Les formes d'apprentissage du Machine Learning varient en fonction de deux facteurs : les données en entrée, également appelées variables explicatives, et les données en sortie, communément nommées cible. Pour chaque forme d'apprentissage, il existe une multitude d'algorithmes spécifiques.

En résumé, le Machine Learning vise à créer des modèles prédictifs à partir de données d'entrée et de sortie, en utilisant différents algorithmes selon la forme d'apprentissage choisie.

On distingue généralement 4 formes d’apprentissage :
- Supervisé
- Semi-supervisé
- Non-supervisé
- Par renforcement

Dans ce module, on se concentrera sur l’**apprentissage supervisé**.

## L'apprentissage supervisé


Dans l'apprentissage supervisé, les données en entrée sont toutes étiquetées ou labellisées, ce qui signifie qu'elles sont associées à une valeur cible ou une catégorie connue. Le modèle apprend alors à faire correspondre les données d'entrée avec les étiquettes correspondantes.

Apprentissage supervisé.

<img src="images/01-supervised-learning.png">

Il existe deux sous-familles parmi les algorithmes d’apprentissage supervisé.
- La classification
- La régression

La classification est une méthode de Machine Learning utilisée lorsque la variable à prédire est catégorielle, c'est-à-dire qu'elle ne peut prendre qu'un ensemble précis de valeurs possibles. Par exemple, la prédiction du genre d'un email (spam ou non-spam) ou la classification d'images en différentes catégories (chien, chat, oiseau, etc.).

Autre Exemples :
- Chat ou chien sur une image
- Survie ou non pour le Titanic
- Variété de fleurs sur les iris



Classification.

<img src="images/02-classification.png">

La régression est une méthode de Machine Learning utilisée lorsque la variable à prédire est continue, c'est-à-dire qu'elle peut prendre une infinité de valeurs possibles. Par exemple, la prédiction de la taille adulte d'un enfant, l'évolution d'une valeur financière ou encore la prédiction du trafic routier.

Pour effectuer une régression, différents modèles peuvent être utilisés, tels que la régression linéaire, la régression polynomiale ou encore la régression logistique.

Pour mettre en place un apprentissage supervisé, trois éléments sont nécessaires :

1. Des données : les données d'entrée $x$ et les labels correspondants $y$. Les données d'entrée peuvent être des images, des textes, des données numériques, etc., tandis que les labels correspondent aux valeurs cibles ou aux catégories associées aux données d'entrée.

2. Un modèle construit sur ces données : le modèle $f(x)$ est utilisé pour faire correspondre une entrée à un label. Il est construit à partir des données d'entrée et des labels correspondants.

3. Une métrique d'évaluation du modèle : la métrique d'évaluation $d(x, f(x))$ est utilisée pour évaluer l'écart entre les prédictions du modèle et la réalité. Elle permet de mesurer la performance du modèle et d'identifier les erreurs éventuelles.


## Evaluer l'apprentissage supervisé

L'évaluation d'un modèle permet de :

1. S'assurer que ses prédictions sont pertinentes : l'évaluation permet de vérifier que les prédictions du modèle sont fiables et correspondent bien à la réalité.

2. Décider de son utilisation ou non : en fonction des résultats de l'évaluation, il est possible de décider si le modèle est suffisamment performant pour être utilisé ou s'il est nécessaire de l'améliorer.

3. Chercher des pistes d'amélioration : l'évaluation permet d'identifier les erreurs éventuelles du modèle et de chercher des moyens de les corriger.

4. Comparer le modèle à d'autres modèles : l'évaluation permet de comparer les performances du modèle à celles d'autres modèles et de choisir le modèle le plus adapté à la tâche à accomplir.

Pour évaluer un modèle, il est nécessaire d'utiliser des métriques adaptées, qui diffèrent en fonction du type de tâche (classification ou régression).

Le principe d'un modèle supervisé est le suivant :

1. Construire un modèle en le faisant apprendre sur des données labellisées : le modèle est entraîné sur un ensemble de données d'entrée pour lesquelles les labels correspondants sont connus.

2. Prendre des données pour lesquelles on connaît le label réel : ces données sont utilisées pour tester le modèle et évaluer sa performance.

3. Les donner au modèle pour qu'il fasse une prédiction du label : le modèle utilise les données d'entrée pour prédire le label correspondant.

4. Comparer le label réel au label prédit : la prédiction du modèle est comparée au label réel pour évaluer la performance du modèle.

## Le surapprentissage

Le surapprentissage se produit lorsque le modèle est trop complexe et ne représente pas fidèlement le phénomène réel.

Surapprentissage.

<img src="images/04-overfitting.png">

Même si un modèle semble donner de bons résultats sur les données d'apprentissage, il peut ne pas se généraliser correctement aux données de test ou aux données réelles. Cela signifie que le modèle peut être trop adapté aux données d'apprentissage et ne pas être capable de prédire correctement les résultats pour des données qu'il n'a pas encore vues.

Pour évaluer la capacité d'un modèle à généraliser, il est important d'utiliser des données de test indépendantes et de mesurer la performance du modèle sur ces données. Il est également important de prendre en compte la complexité du modèle et d'utiliser des techniques de régularisation pour éviter le surapprentissage.

## Entraînement et test

Les données d'entraînement sont utilisées pour entraîner le modèle et lui permettre d'apprendre à partir des données disponibles. Les données de test sont cachées au modèle et sont utilisées uniquement pour évaluer la performance du modèle.

Pour diviser les données disponibles en deux sous-ensembles, on utilise généralement une division aléatoire, avec environ 80% des données réservées pour l'entraînement et 20% pour le test. Cette division permet d'évaluer la capacité du modèle à prédire correctement les résultats pour des données inconnues.

On évalue ensuite la généralisation du modèle:

1. On prend les modèles qui ont appris sur les données d’entraînement
2. On évalue ces modèles sur les données de test

Sur les données de test, il y a une
complexité optimale du modèle. Les données sont à la fois assez complexe mais pas trop pour bien généraliser sur des données inconnues.

## La validation croisée

La méthode la plus courante consiste à diviser les données de manière aléatoire en deux sous-ensembles, avec une taille donnée pour chaque sous-ensemble (généralement 80% pour l'entraînement et 20% pour le test). Il est également possible de prendre en compte la répartition des classes entre les deux sous-ensembles pour s'assurer que chaque classe est représentée de manière équilibrée.

Pour aller plus loin, on peut utiliser des techniques de validation croisée pour évaluer la performance du modèle. La validation croisée consiste à diviser les données en plusieurs jeux d'entraînement et de test, à entraîner le modèle sur chaque jeu d'entraînement et à évaluer sa performance sur chaque jeu de test. Ensuite, on calcule un score moyen pour évaluer la performance globale du modèle.

Validation croisée.

<img src="images/06-cross-validation.png">

#### Séparation des jeux de données avec Scikit-learn

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier

url = 'iris.csv'
iris = pd.read_csv(url)
iris

# Données
X = iris.drop('variety', axis=1)

# Labels
y = iris['variety']
y

# on separe nos donées d'entrainement et nos données de test 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20)

print("Taille de données initials :", len(iris))
print('Nombre de données d\'entrainement', len(X_train), ", propotion", len(X_train)*100/len(iris), '%')
print("Nombre de données de test :", len(X_test), ", proportion", len(X_test)*100/len(iris), '%')

model = DecisionTreeClassifier()

# J'entraine mon model
model.fit(X_train, y_train)
y_pred = model.predict(X_test)

model.score(X_test, y_test)
```

## Les métriques de classification

La validation croisée permet de personnaliser la taille des jeux d'entraînement et de test en fonction des besoins spécifiques de l'application. Par exemple, on peut utiliser une taille de feuille plus petite pour les jeux de données plus volumineux, ce qui permet de réduire le temps de calcul. De plus, la validation croisée permet d'utiliser des méthodes telles que "seul contre tous" en donnant une taille de feuille égale à la taille des données.

La validation croisée offre également la possibilité de définir le score (métrique) utilisé pour évaluer la performance du modèle.En fonction de l'application, différents scores peuvent être utilisés pour évaluer le modèle comme la **précision** ou le **taux d'erreur** en utilisant les notions **vrais positifs** (TP), **vrais négatifs** (TN), **faux positifs** (FP) et **faux négatifs** (FN).

$$
\text{justesse} = \frac{TP + TN}{TP + TN + FP + FN} \quad \quad \text{taux d'erreur} = \frac{FP + FN}{TP + TN + FP + FN}
$$
$$
\text{précision} = \frac{TP}{TP + FP} \quad \quad \text{rappel} = \frac{TP}{TP + FN}
$$

Il existe aussi des métriques pour évaluer la qualité d’un modèle de régression. On utilise principalement l’**erreur quadratique moyenne** (mean squared error, MSE).

$$
MSE = \frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2
$$

Où $N$ est le nombre de données évaluées, $y$ la valeur réelle de la donnée que l’on cherche à prédire et $\hat{y}$ est la valeur prédite par le modèle.

Elle représente la variance moyenne du modèle par rapport à la réalité.

## Classification : L'arbre de décision

Un arbre de décision est un modèle de Machine Learning utilisé pour la classification et la régression. Il peut être vu comme une liste de questions à choix binaire (oui/non) qui permettent de parvenir à une prédiction.

Le fonctionnement d'un arbre de décision est similaire à celui du "jeu des animaux". À chaque étape, l'arbre pose une question sur une caractéristique des données d'entrée et divise les données en fonction de la réponse. Ce processus se répète jusqu'à ce qu'une prédiction finale soit obtenue.

Pour construire un arbre de décision en se basant sur des données, on suit généralement les étapes suivantes :

1. Déterminer une liste de questions pertinentes : Les questions doivent être choisies de manière à ce qu'elles permettent de séparer les données en fonction des classes cibles. Les questions peuvent être basées sur les caractéristiques des données d'entrée, telles que l'âge, le sexe, le revenu, etc.
    
2. Réussir à séparer les données dans les différentes classes : À chaque étape, l'arbre de décision divise les données en fonction de la réponse à la question posée. L'objectif est de séparer les données de manière à ce que chaque feuille de l'arbre contienne des données appartenant à une seule classe cible.

3. Utiliser le moins de questions possible : L'objectif est de construire un arbre de décision aussi simple que possible tout en maintenant une précision de prédiction élevée. Pour ce faire, on essaye d'utiliser le moins de questions possible pour séparer les données en fonction des classes cibles.


Le principe des algorithmes d'arbre de décision est le suivant :

1. Assigner toutes les données d'entraînement à la racine de l'arbre : Toutes les données d'entraînement sont assignées à la racine de l'arbre de décision.

2. Trouver la question qui sépare le mieux les deux classes : L'algorithme cherche la question qui permet de séparer le mieux les données en fonction des classes cibles. Cette question est sélectionnée en utilisant une métrique telle que le gain d'information ou l'indice de Gini.

3. Partitionner les données en fonction de la question : Les données sont partitionnées en fonction de la réponse à la question sélectionnée.
 
4. Si une partition contient des données d'une seule classe : on obtient une feuille de l'arbre : Si toutes les données d'une partition appartiennent à la même classe cible, cette partition devient une feuille de l'arbre de décision.

5. Sinon : on recommence sur chaque nœud qui n'est pas une feuille : Si les données d'une partition appartiennent à plusieurs classes cibles, l'algorithme recommence le processus de sélection de question et de partitionnement sur chaque nœud qui n'est pas une feuille.

Ce processus est répété jusqu'à ce que toutes les données soient classées dans des feuilles de l'arbre de décision. L'arbre de décision final peut être utilisé pour faire des prédictions sur de nouvelles données en suivant le chemin de questions et de réponses jusqu'à atteindre une feuille de l'arbre.

Arbre de décision.

<img src="images/08-decision-tree.png">

Le choix de la meilleure question pour séparer les données dans un arbre de décision dépend de la mesure d'impureté utilisée. Voici les étapes générales pour choisir la meilleure question :

1. Considérer un grand nombre de questions possibles : L'algorithme considère un grand nombre de questions possibles pour diviser les données. Ces questions peuvent être basées sur les caractéristiques des données d'entrée.
    
2. Regarder le partitionnement que l'on obtiendrait : Pour chaque question, l'algorithme regarde le partitionnement des données qui résulterait de la division en fonction de la réponse à cette question.
    
3. Calculer un score d'impureté : Pour chaque partition résultant de la division, l'algorithme calcule un score d'impureté qui mesure le degré de mélange des classes cibles dans la partition. Il existe plusieurs mesures d'impureté, telles que l'entropie, l'indice de Gini, et le gain d'information.
    
4. Choisir la question qui minimise l'impureté : L'algorithme choisit la question qui divise les données de manière à minimiser le score d'impureté moyen sur toutes les partitions résultantes.

Il existe de nombreuses mesures d’impureté. Le coefficient de Gini est un des plus connus.
$$
\text{gini(noeud)} = 1 - \sum_{i=1}^{n} p_i^2
$$

Où $n$ représente le nombre de classes et $p_{i}$ est la probabilité qu'une donnée appartienne à la classe $i$.

Exemple pour la racine de l'arbre de décision.

$$
\text{gini(racine)} = 1 - ((\frac{10}{15})^{2} + (\frac{5}{15})^{2}) = 0.444
$$

Ce coefficient va nous permettre de déterminer les meilleures questions à poser pour séparer les données.

## La régression linéaire

L’objectif de la régression linéaire est de trouver une fonction $f$ telle que $f(x) = y$, sous forme d’un polynôme.

$$
f(x) = w_{0} + \sum_{i=1}^{n} w_{i}x_{i}
$$

Où $w_{i}$ sont les coefficients de la régression, que le modèle cherche à apprendre, en minimisant l’erreur et $x_{i}$ sont les valeurs des données des attributs.

La régression linéaire est un modèle simple et efficace pour prédire des valeurs continues à partir de données d'entrée. Elle est souvent utilisée pour modéliser des relations linéaires entre les variables et pour faire des prédictions sur des données numériques. C'est un modèle simple qui n'est pas toujours adapté.

La régression linéaire.

<img src="images/10-linear-regression.png">

Pour aller plus loin, on peut utiliser des modèles de régression plus complexes, tels que la régression polynomiale. La régression polynomiale est une technique de régression qui permet de modéliser des relations non linéaires entre les variables. Elle consiste à élever les variables à des degrés supérieurs et à combiner les variables entre elles pour obtenir un modèle polynomial.

Par exemple, pour des données avec deux attributs $x_{1}$ et $x_{2}$, une régression polynomiale de degré 2 peut être représentée sous la forme suivante :

$$
f(x) = w_{0} + w_{1}x_{1} + w_{2}x_{2} + w_{3}x_{1}^{2} + w_{4}x_{1}x_{2} + w_{5}x_{2}^{2}
$$

La régression polynomiale permet de modéliser des relations complexes entre les variables, mais elle peut également entraîner un surajustement des données si le degré du polynôme est trop élevé. Il est donc important de choisir le degré approprié en utilisant des techniques telles que la validation croisée.

La régression polynomiale.

#### La régression linéaire avec Scikit-learn

On commence par générer des données qui suivent une distribution connue (cosinus) (dans la réalité, on ne connaîtra bien entendu pas cette vraie distribution : on cherche justement à la modéliser avec la regression linéaire).

```python
# On simule une fonction
def function_simulation(x):
    return np.cos(1.5*np.pi*x)

abscisse = np.arange(0,10, 0.01)
plt.scatter(abscisse, function_simulation(abscisse))
plt.title('Fonction simulée')
plt.xlabel('x')
plt.ylabel('f(x)')
```

```python
np.random.seed(0)

# On prevoit un echantillon de 30 donées
n_samples = 30

X = np.sort(np.random.rand(n_samples))
y = function_simulation(X) + np.random.rand(n_samples) * 0.1

# On affiche les donées réelles qui viennet d'être crées
plt.scatter(X,y)

# On affiche la fonction  a modeliser
#plt.plot(X, function_simulation(X), label="Fonction a modeliser", color="y")
#plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

Maintenant que nous avons des points de simulation, nous allons entraîner un modèle de régression linéaire sur ces données.

```python
X_train, X_test, y_train, y_test = train_test_split(X.reshape(-1, 1), y, test_size=0.4) 
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

# On crée le modele de regression lineaire
linear_regression_model= LinearRegression()
linear_regression_model.fit(X_train, y_train)

# Prediction de label test
y_pred = linear_regression_model.predict(X_test)
print("Erreur quadratique moyenne :", mean_squared_error(y_test, y_pred))
```

Notre modèle est entraîné et nous avons évalué sa performance sur les données de test. Nous pouvons maintenant l'utiliser pour faire des prédictions sur de nouvelles données et visualiser le modèle obtenu.

```python
np.random.seed(0)

# On prevoit un echantillon de 30 donées
n_samples = 30

X = np.sort(np.random.rand(n_samples))
y = function_simulation(X) + np.random.rand(n_samples) * 0.1

X_regression = np.linspace(0, 1, 100)


# On affiche les donées réelles qui viennet d'être crées
plt.scatter(X,y)

# On affiche la fonction  a modeliser
plt.plot(X, function_simulation(X), label="Fonction a modeliser", color="y")
plt.plot(X_regression, linear_regression_model.predict(X_regression.reshape(-1,1)), color="r", label="Model")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```

Dans le cas d'une régression linéaire, les valeurs des différents attributs peuvent avoir un impact différent sur les prédictions du modèle. Si les valeurs d'un attribut sont beaucoup plus élevées que celles d'un autre attribut, cela peut entraîner une surpondération de cet attribut dans les prédictions.

Dans l'exemple donné, une petite variation sur l'attribut $x_{1}$ aura un impact beaucoup plus important sur les prédictions qu'une grande variation sur l'attribut $x_{2}$. C'est pourquoi la normalisation des données peut être nécessaire pour améliorer les performances d'un modèle de régression.

La normalisation consiste à transformer les valeurs des attributs pour qu'elles soient toutes comprises dans une même échelle, généralement entre 0 et 1. Cela permet de donner une importance égale à tous les attributs dans les prédictions du modèle.
