## Matplotlib

```python
import matplotlib.pyplot as plt

plt.plot(iris['sepal.length'], iris['sepal.width'], marker='o', linestyle='', color='m')
plt.xlabel('Longueur du sepal')
plt.ylabel('Largeur du sepal')
plt.title('Largeur du sepal en fonction de la longueur du sepal')
```

```python
iris_setosa = iris[iris['variety']== 'Setosa']
iris_virginica = iris[iris['variety']== 'Virginica']
iris_versicolor = iris[iris['variety']== 'Versicolor']

plt.plot(iris_setosa['sepal.length'], iris_setosa['sepal.width'], marker='1', linestyle='', color='m', label='Setosa')
plt.plot(iris_virginica['sepal.length'], iris_virginica['sepal.width'], marker='+', linestyle='', color='k', label='Virginica')
plt.plot(iris_versicolor['sepal.length'], iris_versicolor['sepal.width'], marker='x', linestyle='', color='r', label='Versicolor')
plt.xlabel('Longueur du sepal')
plt.ylabel('Largeur du sepal')
plt.legend()
plt.title('Largeur du sepal en fonction de la longueur du sepal')
```

```python
fig, axes = plt.subplots(1, 2, figsize=(10, 5))

subplot1 = axes[0]
subplot1.plot(iris['sepal.width'], iris['sepal.length'], marker='1', linestyle='', color='m')
subplot1.set_xlabel('Longueur du sepal')
subplot1.set_ylabel('Largeur du sepal')
subplot1.set_title('Longueur du sepal en fonction de la largeur du sepal')

subplot2 = axes[1]
subplot2.plot(iris['petal.width'], iris['petal.length'], marker='1', linestyle='', color='m')
subplot2.set_xlabel('Longueur du petal')
subplot2.set_ylabel('Largeur du petal')
subplot2.set_title('Longueur du patal en fonction de la largeur du petal')
```
